Faça um programa que leia algo pelo teclado e mostre na tela o seu tipo primitivo e todas as informações possíveis sobre ele.

```python
# DESAFIO 004
palavra = input("Digite uma palavra: ")

print('O tipo primitivo desse valor é', type(palavra))
print('Só tem espaço?', palavra.isspace())
print('É um número?', palavra.isnumeric())
print('É alfabético?', palavra.isalpha())
print('É alfanumérico?', palavra.isalnum())
print('Está em maiúsculas?', palavra.isupper())
print('Está em minúscula?', palavra.islower())
print('Está capitalizada?', palavra.istitle())
```

Crie um algoritmo que leia um número e mostre o seu dobro, triplo e raiz quadrada.

```python
# DESAFIO 006
numero = float(input("Digite um número: "))

dobro = numero * 2
triplo = numero * 3
raiz_quadrada = numero ** 0.5

print(f"O dobro de {numero} é {dobro}")
print(f"O triplo de {numero} é {triplo}")
print(f"A raiz quadrada de {numero} é {raiz_quadrada:.4f}")
```

Faça um programa que leia a largura e a altura de uma parede em metros, calcule a sua área e a quantidade de tinta necessária para pintá-la, sabendo que cada litro de tinta pinta uma área de 2 metros quadrados.

```python
# DESAFIO 011
largura = float(input("Largura da parede:"))
altura = float(input("Altura da parede:"))

area = largura * altura
litroTinta = area / 2

print(f'Sua parede tem a dimensão de {largura:.2f} X {altura:.2f} e sua área é de {area}m².')
print(f'Para pintar essa parede você vai precisar de {litroTinta:.2f}L de tinta.')
```

Faça um algoritmo que leia o preço de um produto e mostre seu novo preço, com 5% de desconto.

```python
# DESAFIO 012
preco_original = float(input("Digite o preço do produto: "))

desconto_percentual = 5
desconto = (desconto_percentual / 100) * preco_original
novo_preco = preco_original - desconto

print(f"O preço original é R${preco_original:.2f}")
print(f"Com {desconto_percentual}% de desconto, o novo preço é R${novo_preco:.2f}")
```

Faça um algoritmo que leia o salário de um funcionário e mostre seu novo salário, com 15% de aumento.

```python
# DESAFIO 013
salario_atual = float(input("Digite seu salário atual: R$ "))

aumento_percentual = 15
aumento = (aumento_percentual / 100) * salario_atual
novo_salario = salario_atual + aumento

print(f"seu salário atual é R${salario_atual:.2f}")
print(f"Com {aumento_percentual}% de aumento, o seu novo salário é R${novo_salario:.2f}")
```

Escreva um programa que converta uma temperatura digitando em graus Celsius e converta para graus Fahrenheit.

```python
# DESAFIO 014
grauCelsius = float(input("Informe a temperatura em °C: "))

grauFahrenheit = (grauCelsius * 9/5) + 32

print(f"A temperatura {grauCelsius}°C equivale a {grauFahrenheit}°F")
```

Escreva um programa que pergunte a quantidade de Km percorridos por um carro alugado e a quantidade de dias pelos quais ele foi alugado. Calcule o preço a pagar, sabendo que o carro custa R$60 por dia e R$0,15 por Km rodado.

```python
# DESAFIO 015
km_percorridos = float(input("Informe a quantidade de Km percorridos: "))
dias_aluguel = int(input("Informe a quantidade de dias de aluguel: "))

custo_por_dia = 60
custo_por_km = 0.15

preco_total = (dias_aluguel * custo_por_dia) + (km_percorridos * custo_por_km)

print(f"O preço a pagar é R${preco_total:.2f}")
```
