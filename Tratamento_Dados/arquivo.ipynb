# TIPOS PRIMITIVOS:

- **int**: Números inteiros, por exemplo: `7`, `4`, `20`, `25`

- **float**: Números quebrados, por exemplo: `3.2`, `5.7`, `9.1`, `4.5`

- **str**: Frases, por exemplo: `'olá'`, `'vois não passarás'`, `'aobaa, bão demais?'`, `''` (string vazia)

- **bool**: Verdadeiro ou falso, por exemplo: `True`, `False`

## Operadores Aritméticos
Os operadores aritméticos são utilizados para realizar operações matemáticas.

| Operador | Descrição               | Exemplo          |
|----------|-------------------------|------------------|
| `+`      | Adição                  | `resultado = 5 + 3` |
| `-`      | Subtração               | `resultado = 5 - 3` |
| `*`      | Multiplicação           | `resultado = 5 * 3` |
| `/`      | Divisão                 | `resultado = 5 / 3` |
| `//`     | Divisão inteira         | `resultado = 5 // 3` |
| `%`      | Resto da divisão        | `resultado = 5 % 3` |
| `**`     | Exponenciação           | `resultado = 5 ** 3` |

## Operadores de Atribuição
Os operadores de atribuição são utilizados para atribuir valores a variáveis.

| Operador | Descrição                | Exemplo              |
|----------|--------------------------|----------------------|
| `=`      | Atribuição               | `x = 5`              |
| `+=`     | Adição e atribuição      | `x += 3` (equivalente a `x = x + 3`) |
| `-=`     | Subtração e atribuição   | `x -= 2` (equivalente a `x = x - 2`) |
| `*=`     | Multiplicação e atribuição | `x *= 4` (equivalente a `x = x * 4`) |
| `/=`     | Divisão e atribuição     | `x /= 2` (equivalente a `x = x / 2`) |

## Operadores de Comparação
Os operadores de comparação são utilizados para comparar valores.

| Operador | Descrição         | Exemplo              |
|----------|-------------------|----------------------|
| `==`     | Igual a           | `resultado = (5 == 5)` |
| `!=`     | Diferente de       | `resultado = (10 != 5)` |
| `<`      | Menor que          | `resultado = (3 < 7)` |
| `>`      | Maior que          | `resultado = (8 > 5)` |
| `<=`     | Menor ou igual a   | `resultado = (4 <= 4)` |
| `>=`     | Maior ou igual a   | `resultado = (6 >= 5)` |

## Operadores Lógicos
Os operadores lógicos são utilizados para realizar operações lógicas.

| Operador | Descrição                     | Exemplo                 |
|----------|-------------------------------|-------------------------|
| `and`    | Retorna True se ambas as condições são verdadeiras | `resultado = (True and False)` |
| `or`     | Retorna True se pelo menos uma das condições é verdadeira | `resultado = (True or False)` |
| `not`    | Inverte o valor lógico (True vira False e vice-versa) | `resultado = not True` |


```python
# EXEMPLO DE INTEIRO

num1 = int(input('Digite o primeiro número'))
num2 = int(input('Digite o segundo número'))

soma = num1 + num2

print(f'A soma vale {soma}')
```

```python
# EXEMPLO DE QUEBRADOS

num1 = float(input('Digite o primeiro número'))
num2 = float(input('Digite o segundo número'))

soma = num1 + num2

print('A soma vale {:.2f}'.format(soma))
```

```python
# EXEMPLO DE STRING

frase = input('Digite algo aqui')

print(f'O tipo primitivo desta variavel é {type(frase)}')
```
