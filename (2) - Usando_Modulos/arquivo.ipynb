# Manipulação de Strings em Python
Em Python, a manipulação de strings é uma parte essencial da programação. A seguir, apresenta-se um resumo em Markdown sobre algumas operações comuns de manipulação de strings:

## Concatenação de Strings
A concatenação de strings envolve a junção de duas ou mais strings em uma única string. Isso é feito usando o operador '+'


````python
# CONCATENAÇÃO

nome = 'São João Batista da Serra Pereira Júnior'
print(f'O nome é:',nome)

nome_completo = nome + " da Silva"

print(f'Adicionando "da Silva" fica:',nome_completo)
````


## Tamanho de uma String
Para obter o tamanho de uma string (o número de caracteres que ela contém), você pode usar a função len().


````python
# TAMANHO DE UMA STRING

nome = 'São João Batista da Serra Pereira Júnior'
print(f'Tem {len(nome)} letras contando espaços')
````


## Acesso a Caracteres
É possível acessar caracteres individuais em uma string usando a indexação. O índice começa em 0 para o primeiro caractere.


````python
# ACESSO A CARACTERES

nome = 'São João Batista da Serra Pereira Júnior'
print(f'Primeira letra é:',nome[0])
````


## Fatiamento de Strings
O fatiamento permite extrair uma parte específica de uma string, especificando uma faixa de índices.



````python
# FATIAMENTO

nome = 'São João Batista da Serra Pereira Júnior'
print(f'Do início até a décima casa fica:', nome[0:10])
````


## Conversão de Maiúsculas e Minúsculas
Os métodos upper() e lower() são usados para converter uma string para maiúsculas e minúsculas, respectivamente.


````python
# CONVERSÃO DE MAIÚSCULAS E MINÚSCULAS

nome = 'São João Batista da Serra Pereira Júnior'

nome_maiusculo = nome.upper()
nome_minusculo = nome.lower()

print(nome_maiusculo)
print(nome_minusculo)
````


## Substituição em uma String
O método replace() substitui uma substring por outra em uma string.


````python
# SUBSTITUIÇÃO EM UMA STRING

nome = 'São João Batista da Serra Pereira Júnior'
nome_substituido = nome.replace("São", "Senhor")
print(nome_substituido)
````


## Busca em uma String
O método find() é usado para encontrar a posição da primeira ocorrência de uma substring em uma string.


````python
# BUSCA EM UMA STRING

nome = 'São João Batista da Serra Pereira Júnior'
posicao_sobrenome = nome.find("Pereira")
print(f'"Pereira" começa na {posicao_sobrenome}° posicão')
````


## Verificação de Substring
A expressão in é usada para verificar se uma substring está presente em uma string.


````python
# VERIFICAÇÃO DE SUBSTRING

nome = 'São João Batista da Serra Pereira Júnior'

verifica_nome = "Batista" in nome
print(f'Tem batista no nome? ', verifica_nome)
````

## Método Split
O método split() divide uma string em uma lista de substrings com base em um separador especificado.


````python
# MÉTODO SPLIT

nome = 'São João Batista da Serra Pereira Júnior'
palavras_do_nome = nome.split(" ")
print(palavras_do_nome)
````


## Método Strip
O método strip() é usado para remover espaços em branco no início e no final de uma string.


````python
# MÉTODO STRIP

nome = 'São João Batista da Serra Pereira Júnior'

nome_com_espacos = "    " + nome + "    "
print(f'Nome com espaços em volta: {nome_com_espacos}')

nome_sem_espacos = nome_com_espacos.strip()
print(f'Nome sem espaços em volta: {nome_sem_espacos}')
````


## Método Count
O método count() conta o número de ocorrências de uma substring em uma string.


````python
# MÉTODO COUNT

nome = 'São João Batista da Serra Pereira Júnior'
contagem_letra_a = nome.count("a")
print(f'Tem {contagem_letra_a} letras "a"')
````


## Método Join
O método join() é utilizado para concatenar elementos de uma lista em uma única string, utilizando uma string de junção como separador.

````python
# MÉTODO JOIN

nome = 'São João Batista da Serra Pereira Júnior'
partes_do_nome = nome.split(" ")

nome_junto = "_".join(partes_do_nome)
print(nome_junto)
````

# Introdução a Módulos e Bibliotecas em Python

Em Python, módulos e bibliotecas são recursos essenciais que permitem organizar e reutilizar código de forma eficiente. Eles oferecem funcionalidades adicionais além das built-ins da linguagem, expandindo as capacidades do Python para uma variedade de aplicações.

## O que são Módulos?

**Módulos** em Python são arquivos contendo definições e instruções Python. Eles podem incluir funções, classes e variáveis. Ao organizar o código em módulos, podemos dividir nossos programas em partes mais gerenciáveis e lógicas.

### Exemplo de Uso de Módulos:

```python
# Importando um módulo
import nome_do_modulo

# Acessando uma função do módulo
nome_do_modulo.funcao()


````python
# EXEMPLO COM MATH
# raiz quadrada
import math

numero = 49

raiz = math.sqrt(numero)

print(raiz)
````
