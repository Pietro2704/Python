Listas sÃ£o uma boa maneira de organizar dados, mas uma desvantagem Ã© que sÃ³ podemos representar valores.

Por que isso Ã© um problema? Por exemplo, alguÃ©m olhando para [115910.26, 128.0, 4] nÃ£o saberia a quais valores correspondem o preÃ§o, Ã¡rea, etc. Uma opÃ§Ã£o melhor pode ser um dicionÃ¡rio, onde cada valor estÃ¡ associado a uma chave.

```python
# DicionÃ¡rio
casa_0 = {
    "PreÃ§o": 115910.26,
    "Ãrea": 128,
    "Quartos": 4,
}

print("casa_0 type:", type(casa_0))

casa_0
```

```python
# Calcular o preco_m2
preco_m2 = casa_0["PreÃ§o"] / casa_0["Ãrea"]
casa_0["PreÃ§o_m2"] = preco_m2

print("casa_0 type:", type(casa_0))

casa_0
```
