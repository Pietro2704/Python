## Introdução
Em programação, as estruturas condicionais permitem que um programa execute diferentes blocos de código com base em condições específicas. Em Python, utilizamos as instruções `if`, `else` e `elif` para implementar a lógica condicional.

## Instrução `if`
A instrução `if` é utilizada para executar um bloco de código se uma condição for verdadeira.

```python
if condicao:
    # Bloco de código a ser executado se a condição for verdadeira


## Instrução else
A instrução else é utilizada para executar um bloco de código se a condição do if for falsa.

```python
if condicao:
    # Bloco de código a ser executado se a condição for verdadeira
else:
    # Bloco de código a ser executado se a condição for falsa

## Instrução elif
A instrução elif (abreviação de "else if") é utilizada para verificar condições adicionais, além da condição do if.

```python
if condicao1:
    # Bloco de código a ser executado se a condição1 for verdadeira
elif condicao2:
    # Bloco de código a ser executado se a condição1 for falsa e a condição2 for verdadeira
else:
    # Bloco de código a ser executado se todas as condições anteriores forem falsas

````python
# EXEMPLO 1
nome = input('Digite seu nome').strip()
nome = nome.capitalize()

if nome != 'Pietro':
  print('NÃO AUTORIZADO')
  
else:
  print(f'Olá {nome}!')


print('Tenha um ótimo dia!')
````

````python
# EXEMPLO 2
numero = int(input("Digite um número: "))

if numero % 2 == 0:
    print("O número é par.")
else:
    print("O número é ímpar.")
````

````python
# EXEMPLO 3
numero = int(input("Digite um número: "))

if numero > 0:
    print("O número é positivo.")
elif numero < 0:
    print("O número é negativo.")
else:
    print("O número é zero.")
````
