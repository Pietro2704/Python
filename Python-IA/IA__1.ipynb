Agora que você já consegue trabalhar com os dados de uma única casa, vamos pensar em como organizar todo o dataset. Uma opção seria criar uma lista para cada observação e, em seguida, juntá-las em outra lista

Isso é chamado de __lista aninhada.__

__dataset__

```python
# Lista Aninhada / Matriz
casas_aninhadas = [
    [115910.26, 128.0, 4.0],
    [48718.17, 210.0, 3.0],
    [28977.56, 58.0, 2.0],
    [36932.27, 79.0, 3.0],
    [83903.51, 111.0, 3.0],
]

print("Tipo da variável:", type(casas_aninhadas))
print("Tamanho da lista:", len(casas_aninhadas))

casas_aninhadas
```

```python
# Acessando elementos da Matriz:
print("Primeiro registro:", casas_aninhadas[0])
print("Preço da primeira casa:", casas_aninhadas[0][0])
```

Listas são uma boa maneira de organizar dados, mas uma desvantagem é que só podemos representar valores.

Por que isso é um problema? Por exemplo, alguém olhando para [115910.26, 128.0, 4] não saberia a quais valores correspondem o preço, área, etc. Uma opção melhor pode ser um dicionário, onde cada valor está associado a uma chave.

```python
# Dicionário
casa_0 = {
    "Preço": 115910.26,
    "Área": 128,
    "Quartos": 4,
}

print("casa_0 type:", type(casa_0))

casa_0
```

```python
# Calcular o preco_m2
preco_m2 = casa_0["Preço"] / casa_0["Área"]
casa_0["Preço_m2"] = preco_m2

print("casa_0 type:", type(casa_0))

casa_0
```

Se quisermos combinar todas as nossas observações, a melhor maneira seria criar uma lista de dicionários.

__dataset__

```python
# Lista de dicionários
casas = [
    {
        "Preço": 115910.26,
        "Área": 128,
        "Quartos": 4,
    },
    {
        "Preço": 48718.17,
        "Área": 210,
        "Quartos": 3,
    },
    {
        "Preço": 28977.56,
        "Área": 58,
        "Quartos": 2,
    },
    {
        "Preço": 36932.27,
        "Área": 79,
        "Quartos": 3,
    },
    {
        "Preço": 83903.51,
        "Área": 111,
        "Quartos": 3,
    },
]

print("casas type:", type(casas))
print("casas length:", len(casas))

casas

# casas[0]["Preço"]
```
